```python
chatlogs_clean = []



for g in chatlogs_:
    #print(g)
    b = []
    for i in g:
        i = i.split()
        i = i[1:]
        i = " ".join(i)
        b.append(i)
        
    g = b
    chatlogs_clean.append(g[2:-2])
```

```python
arg_dic = {}

for i in chatlogs_clean:
    it = iter(i)
    tuple_log = list(zip(it,it))
    for tup in tuple_log:
        arg_id = tup[0]
        user_ca = tup[1]
        if arg_id not in arg_dic:
            arg_dic[arg_id] = [user_ca]
        else:
            arg_dic[arg_id].append(user_ca)
```
